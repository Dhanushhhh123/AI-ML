```python
import pandas as pd
```

```python
dataset=pd.read_csv("/content/Weather Dataset.csv")
```

```python
dataset["Weather"] = dataset["Weather"].str.lower()
```

```python
dataset["Weather_Simplified"] = dataset["Weather"].replace({
    r".(clear).": "Clear",                 # clear, mainly clear
    r".(cloud).": "Cloud",                 # cloudy, mostly cloudy
    r".(rain|drizzle|shower|thunderstorm).": "Rain",  # rain, drizzle, thunderstorm, showers
    r".(snow|flurries|pellet).": "Snow",   # snow, snow showers, pellets
    r".(fog|mist|haze).": "Fog"            # fog, mist, haze
}, regex=True)

# Check distribution
print(dataset["Weather_Simplified"].value_counts())

```

```python
def map_weather_keywords(w):
    w = w.lower()
    if 'clear' in w:
        return 'Clear'
    elif 'cloudy' in w:
        return 'Cloudy'
    elif 'rain' in w or 'drizzle' in w or 'shower' in w:
        return 'Rain'
    elif 'snow' in w or 'pellet' in w:
        return 'Snow'
    elif 'fog' in w or 'haze' in w:
        return 'Fog/Mist/Haze'
    elif 'thunder' in w:
        return 'Thunderstorm'
    elif 'freezing' in w or 'ice' in w:
        return 'Ice/Freezing'
    else:
        return 'Other'

dataset['Weather_Category'] = dataset['Weather'].apply(map_weather_keywords)
display(dataset['Weather_Category'].value_counts())
```

```python
# Check skewness for all numeric columns
dataset.skew(numeric_only=True)
import seaborn as sns
import matplotlib.pyplot as plt

for col in dataset.select_dtypes(include=['int64','float64']).columns:
    plt.figure(figsize=(6,4))
    sns.histplot(dataset[col], kde=True)
    plt.title(f"Distribution of {col}")
    plt.show()
```
